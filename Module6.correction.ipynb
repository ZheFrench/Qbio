Install matplotlib-venn.

```python
import pandas as pd

file = "./data/set1.txt"
data = pd.read_csv(file,sep='\t',header=0)
data.head()
list1 = data['Set1'].tolist()

file = "./data/set2.txt"
data = pd.read_csv(file,sep='\t',header=0)
data.head()
list2 = data['Set2'].tolist()

# Import libraries
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib.patches import Circle
from matplotlib import pyplot as plt
import numpy as np

list1_set = set(list1)
list2_set = set(list2)

# Ven diagram
v = venn2([list1_set, list2_set], set_labels=("set1", "set2"),set_colors=('mediumaquamarine', 'slategrey'), alpha = 0.5)
venn2_circles(subsets=(len(list1_set - list2_set), 
                       len(list2_set - list1_set), 
                       len(list1_set.intersection(list2_set))), linewidth=1, color="orange")
plt.title("Ven diagram ")
plt.annotate('Set1', xy=v.get_label_by_id('10').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))

plt.annotate('Set2', xy=v.get_label_by_id('11').get_position() + np.array([0, 0.05]), xytext=(70,70),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.4',color='gray'))
```
