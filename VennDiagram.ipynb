```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the file
path_3= 'D:\\Users\\Dazai\\Documents\\BON\\Data_visualisation\\Question_1\\cohort1_All_Pre_1_default_vs_relaxed.tsv'
df3 = pd.read_csv(path_3, delimiter='\t')

# Extract the columns
data_x = df3['qval.x'].dropna()
data_y = df3['qval.y']

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].boxplot(data_x)
axs[0].set_title('qval.x')
axs[1].boxplot(data_y)
axs[1].set_title('qval.y')
plt.show()
```

```python
## What are the interactions that fail to be captured by the bulk approach ? Describe them. (cohort1_All_Pre_1_default_vs_relaxed.tsv)
import matplotlib.pyplot as plt

# Extract the columns
x_corr = df3['LR.corr.x'].dropna()
x_qval = df3['qval.x'].dropna()
y_corr = df3['LR.corr.y'].dropna()
y_qval = df3['qval.y'].dropna()


# Create a figure
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

# Create the scatter plot

ax.scatter(y_corr, y_qval, color='green', label='y')
ax.scatter(x_corr, x_qval, color='orange', label='x')

# Add a legend
ax.legend()

# Set the x-axis label
ax.set_xlabel('LR Correlation')

# Set the y-axis label
ax.set_ylabel('qval')

# Show the plot
plt.show()
```
